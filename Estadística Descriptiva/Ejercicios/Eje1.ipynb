1. Media (promedio)

Un investigador registra el número de horas que 8 estudiantes estudiaron para un examen:
[2, 4, 3, 5, 6, 2, 4, 8]
📌 Ejercicio: Calcula la media de horas estudiadas.

```python
import numpy as np
```

```python
arreglo1 = [2, 4, 3, 5, 6, 2, 4, 8]

media = sum(arreglo1) / len(arreglo1)
print("Media:", media)
```

2. Mediana

Una encuesta registra la cantidad de cafés bebidos en un día por 7 personas:
[1, 0, 2, 3, 1, 2, 4]
📌 Ejercicio: Calcula la mediana.

```python
arreglo2 = np.array([1, 0, 2, 3, 1, 2, 4])
print("Arreglo original", arreglo2)
print(f'Arreglo Ordenado: {np.sort(arreglo2)}')
print(f'Mediana: {np.median(arreglo2)}')
```

3. Desviación estándar

Un sensor mide temperaturas durante 6 días:
[18, 21, 19, 22, 20, 24]
📌 Ejercicio: Calcula la desviación estándar.

```python
arreglo3 = np.array([18, 21, 19, 22, 20, 24])
print(f'Media = {np.mean(arreglo3)}, s = {np.std(arreglo3)}, var = {np.var(arreglo3)}')
```

4. Rango intercuartílico (IQR)

Los ingresos diarios de 10 tiendas son:
[1200, 1500, 1600, 1400, 1550, 1700, 1800, 1450, 1650, 1750]
📌 Ejercicio: Calcula el rango intercuartíl (Q3 − Q1).

```python
arreglo4 = np.array([1200, 1500, 1600, 1400, 1550, 1700, 1800, 1450, 1650, 1750])

print(f'Arreglo original: {arreglo4}')
print(f'Arreglo ordenado: {np.sort(arreglo4)}')

q1 = np.percentile(arreglo4, 25)
q3 = np.percentile(arreglo4, 75)

print(f'Q3: {q3}, Q1: {q1}, Rango Intercuartílico: {q3 - q1}')
```

🔹 5. Moda (categorías)

En una empresa se registran los colores de autos vendidos durante un mes:
[Rojo, Azul, Blanco, Blanco, Azul, Blanco, Negro, Azul, Azul]
📌 Ejercicio: Encuentra la moda.

```python
import numpy as np

arreglo = np.array([4, 4, 5, 6, 6, 6, 7, 8])

media = np.mean(arreglo)
mediana = np.median(arreglo)
desviacion = np.std(arreglo)

# Moda manual con NumPy
valores, cuentas = np.unique(arreglo, return_counts=True)
moda = valores[np.argmax(cuentas)]

print("Media:", media)
print("Mediana:", mediana)
print("Desviación estándar:", desviacion)
print("Moda:", moda)
```

6. Conteo de categorías (frecuencia absoluta)

Una encuesta pregunta el tipo de mascota favorita:
[Perro, Gato, Perro, Ave, Perro, Gato, Pez, Perro]
📌 Ejercicio: Calcula cuántas veces aparece cada categoría.

```python
from collections import Counter

mascotas = ["Perro", "Gato", "Perro", "Ave", "Perro", "Gato", "Pez", "Perro"]

conteo = Counter(mascotas)

print(conteo)
```
