1. Media (promedio)

Un investigador registra el número de horas que 8 estudiantes estudiaron para un examen:
[2, 4, 3, 5, 6, 2, 4, 8]
📌 Ejercicio: Calcula la media de horas estudiadas.

```python
import numpy as np
```

```python
arreglo1 = [2, 4, 3, 5, 6, 2, 4, 8]

media = sum(arreglo1) / len(arreglo1)
print("Media:", media)
```

2. Mediana

Una encuesta registra la cantidad de cafés bebidos en un día por 7 personas:
[1, 0, 2, 3, 1, 2, 4]
📌 Ejercicio: Calcula la mediana.

```python
arreglo2 = np.array([1, 0, 2, 3, 1, 2, 4])
print("Arreglo original", arreglo2)
print(f'Arreglo Ordenado: {np.sort(arreglo2)}')
print(f'Mediana: {np.median(arreglo2)}')
```

3. Desviación estándar

Un sensor mide temperaturas durante 6 días:
[18, 21, 19, 22, 20, 24]
📌 Ejercicio: Calcula la desviación estándar.

```python
arreglo3 = np.array([18, 21, 19, 22, 20, 24])
print(f'Media = {np.mean(arreglo3)}, s = {np.std(arreglo3)}, var = {np.var(arreglo3)}')
```
