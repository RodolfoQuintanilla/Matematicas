```python
# Leer dataset
import pandas as pd
import numpy as np

df = pd.read_csv("tasas-crimenes.csv", encoding="latin-1")

df.head()
```

```python
print(f'Promedio tasa de asesinatos: {df["Murder.Rate"].mean()}')
print(f'Media tasa de asesinatos: {df["Murder.Rate"].median()}')
```

```python
df.head()
```

```python
df['Nro_asesinatos'] = (1/1000000) * df['Murder.Rate']*df['Population']
df.head()
```

```python
print(f"Promedio de asesinatos: {df['Nro_asesinatos'].mean()}")
print(f"Mediana de asesinatos: {df['Nro_asesinatos'].median()}")
```

```python
df.describe()
```

```python
print('Población:')
print(f"\tMedia: {df['Population'].mean()}")
print(f"\tDesviación: {df['Population'].std()}")

print('='*30)
print('Tasa de asesinatos:')
print(f"\tMedia: {df['Murder.Rate'].mean()}")
print(f"\tDesviación: {df['Murder.Rate'].std()}")
```

```python
pr = [25, 50, 75]
cuartiles = np.percentile(df['Population'], pr)

print(f'Arreglo organizado: {np.sort(df["Population"])}')
for p, pv in zip(pr, cuartiles):
    print(f"Percetil {p}: {pv}")
```
