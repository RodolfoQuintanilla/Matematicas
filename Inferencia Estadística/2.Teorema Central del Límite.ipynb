# Teorema Central del Límite (TCL)

🎯 Objetivo

Entender por qué las medias de muestras se vuelven normales, incluso si los datos originales no lo son.

🧠 1. Idea clave (en simple)

El:

Teorema Central del Límite

dice:

Si tomas muchas muestras de una población y calculas sus medias, esas medias siguen una distribución normal.

👉 Aunque los datos originales NO sean normales.

🔥 Ejemplo intuitivo

Imagina:

Datos originales → totalmente desordenados (uniformes, raros, etc.)
Tomas muchas muestras
Calculas sus medias

👉 Resultado:
Una campana perfecta 😮

```python
import numpy as np

datos = np.random.uniform(low=0, high=100, size=10000)
```

```python
datos
```

👉 Esto es una distribución uniforme (no campana).

🧪 Paso 2: Tomar muestras y calcular medias

```python
medias = []

for i in range(1000):
    muestra = np.random.choice(datos, size=30)
    medias.append(np.mean(muestra))
```

```python
medias
```

🧪 Paso 3: Graficar

```python
import matplotlib.pyplot as plt

plt.hist(medias, bins=50, edgecolor='white')
plt.title("Distribución de medias")
plt.show()
```

🧠 3. Lo que debes observar

👉 Aunque los datos originales eran uniformes:

La gráfica de medias se ve como:
✔ una campana
✔ centrada en la media
