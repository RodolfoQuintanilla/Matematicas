```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Ejercicio 1: Generar datos normales

Objetivo:
Simular datos como si fueran reales.

Instrucciones:
Genera 1000 datos con:
media = 50
desviación = 10
Guárdalos en una variable

```python
datos = np.random.normal(loc=50, scale=10, size=1000)
```

```python
print(datos)
```

## Ejercicio 2: Analizar los datos

Calcula:

media
desviación estándar

👉 Pregunta clave:

¿Se parecen a los valores que definiste?

```python
# Media

mediaDatos = np.mean(datos)
print(f'Media de Datos es: {mediaDatos}')
```

```python
desviacionEstandarDatos =np.std(datos)
print(f"Desviacion estandar: {desviacionEstandarDatos}")
```

## Ejercicio 3: Visualización

Haz un histograma.

👉 Aquí ocurre la magia:

¿Se ve como campana?
¿Está centrado en 50?

```python
# Histograma.

plt.hist(datos, bins=50, color='red', edgecolor='white')
plt.title("Histograma de Alturas")
plt.xlabel("Altura (cm)")
plt.ylabel("Frecuencia")
plt.show()
```
